## Основные теоретические положения

**Ранжированный ряд** – это распределение отдельных единиц совокупности в порядке возрастания или убывания исследуемого признака. Ранжирование позволяет легко разделить количественные данные по группам, сразу обнаружить наименьшее и наибольшее значения признака, выделить значения, которые чаще всего повторяются.

**Вариационный ряд** – последовательность значений заданной выборки $ x^m=(x_1,…,x_m) $, расположенных в порядке неубывания:
$$x^{(1)} \leqslant x^{(2)} \leqslant ... \leqslant x^{(m)} $$

**Интервальный ряд распределения** – это таблица, состоящая из двух столбцов (строк) – интервалов варьирующего признака $X_i$ и числа единиц совокупности, попадающих в данный интервал (частот - $f_i$), или долей этого числа в общей численности совокупностей (частостей - $d_i$).

**Полигоном частот** называют ломанную, отрезки которой соединяют точки $(x_1,n_1), (x_2,n_2), ...,(x_k,n_k)$. Для построения полигона частот на оси абсцисс откладывают варианты $x_i$, а на оси ординат – соответствующие им частоты $n_i$. Точки $(x_i,n_i)$ соединяют отрезками прямых и получают полигон частот.

**Гистограммой частот** (частостей) называется ступенчатая фигура, состоящая из прямоугольников с основаниями, равными интервалам значений hi и высотами, равными отношению частот (или частостей) к шагу:
$$\dfrac{m_i}{h_i}(\dfrac{\omega_i}{h_i} = \dfrac{m_i}{n \cdot h_i})$$

Эмпирической функцией распределения, построенной по выборке $x^m = (x_1, ..., x_m)$  объема $m$, называется случайная функция $\hat{F}_m(x)$, равная $$\hat{F}_m(x) = \dfrac{1}{m} \sum\limits^{m}_{i = 1}I_{\{x_i \leqslant x\}}$$ Значения эмпирической функции распределения принадлежат отрезку $[0, 1]$.

Для определения количества интервалов используем формулу Стерджесса: 
$$k = 1 + 3.31 \cdot \lg n$$
где $n$ – объем выборки.
Используя в качестве $n = 107$, получаем, что $k = 7$.
Чтобы определить шаг, с которым формировать интервалы, использована формула: 
$$ h = \frac{x_{\max} - x_{\min}}{k}$$
Соответственно, для $x_{\min} = 320$, $x_{\max} = 593$ и $k = 8$ получаем, что $h \approx 39$.

```python
k = 1.33 + 3.31 * np.log10(len(df))
k = int(k)
h = (max(ranked_series) - min(ranked_series)) / k
h = int(h)
means = []
nums = []
relative_nums = []
distrib_nums = []
d1 = []
for i in range(min(ranked_series), max(ranked_series), h):
    cond = (i <= variation_df['Значение']) & (variation_df['Значение'] < i+h) if i+h < max(ranked_series) else (i <= variation_df['Значение']) & (variation_df['Значение'] <= i+h)
    means.append(int(variation_df['Значение'][cond].mean()))
    nums.append(variation_df['Частота'][cond].sum())
    d1.append(variation_df['Частота'][variation_df['Значение'] <= i+h].sum())
    relative_nums.append(variation_df['Относительная частота'][cond].sum())
    distrib_nums.append(variation_df['Относительная частота'][variation_df['Значение'] <= i+h].sum())
```

В сумме абсолютные частоты дают 107, что соответствует объему выборки, а относительные частоты суммируются к единице.

```python
inter_df = pd.DataFrame({'Средние значения': means, 'Частота': nums}, dtype=np.int64)
inter_df['Частота'] = inter_df['Частота'] / h

relative_inter_df = pd.DataFrame({'Средние значения': means, 'Относительная частота': relative_nums}, dtype=np.int64)


fig = inter_df.plot(x='Средние значения', y='Частота', title='Полигон для абсолютной частоты')
plt.grid()
plt.show()

fig = inter_df.plot(x='Средние значения', y='Частота', kind='bar', title='Гистограмма для абсолютной частоты')
plt.grid()
plt.show()

plt.step(x = means, y =d1, where='mid')
plt.xlabel('Средние значения')
plt.ylabel('Частота')
plt.title('Эмпирическая функция для абсолютной частоты')
plt.grid()
plt.show()

fig = relative_inter_df.plot(x='Средние значения', y='Относительная частота', title='Полигон для относительной частоты')
plt.grid()
plt.show()

fig = relative_inter_df.plot(x='Средние значения', y='Относительная частота', kind='bar', title='Гистограмма для относительной частоты')
plt.grid()
plt.show()

plt.step(x=means, y=distrib_nums, where='mid')
plt.xlabel('Средние значения')
plt.ylabel('Относительная частота')
plt.title('Эмпирическая функция для относительной частоты')
plt.grid()
plt.show()
```
