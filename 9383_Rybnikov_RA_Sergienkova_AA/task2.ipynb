## Основные теоретические положения

$\bullet$ Начальным эмпирическим моментом $k$-го порядка называется среднее значение $k$-х степеней элементов вариационного или интервального ряда. Вычисляется по формуле:

$$\displaystyle \overline M_k = \dfrac{1}{N}\sum n_jx_j^k$$

В частности

$$\displaystyle \overline x_в = \overline M_1 = \dfrac{1}{N}\sum n_jx_j$$


$\bullet$ Центральным эмпирическим момент $k$-го порядка называется среднее значение $k$-x степеней разностей $x_j - \hat x_в$ для вариационного или интервального ряда. Вычисляется по формуле:

$$\displaystyle \overline m_k = \dfrac{1}{N}\sum n_j\cdot(x_j-\overline x_в)^k$$

В частности

$$\displaystyle \overline D_в = \overline m_2 = \dfrac{1}{N}\sum n_j\cdot(x_j-\overline x_j)^2$$

$\bullet$ Это не смещенная оценка, поэтому следует использовать исправленную оценку дисперсии:

$$\displaystyle s^2 = \dfrac{N}{N-1}D_в$$

$\bullet$ Статистические оценки ассиметрии вычисляются по формуле:

$$\displaystyle \overline A_s = \dfrac{\overline m_3}{s^3}$$

$\bullet$ Статистические оценки эксцесса вычисляются по формуле:

$$\displaystyle \overline E = \dfrac{\overline m_4}{s^4}-3$$

$\bullet$ Условные варианты вычисляются по формуле:

$$\displaystyle u_j = \dfrac{x_j - C}{h},$$

где $\displaystyle C$ - условный ноль, значение которого выбирается равным значению варианты интервального ряда, являющуюся средней или близкой к средней по значению в этом ряду.

$\bullet$ Условные моменты $k$-го порядка вычисляются по формуле:

$$\displaystyle \overline M_k^* = \dfrac{1}{N}\sum n_ju_j^k$$

Связь условны эмпирических моментов с эмпирическими начальными и центральными моментами:

$$\displaystyle \overline x_в = \overline M_1^*h + C = \overline M_1$$

$$\displaystyle \overline m_2 = \Big(\overline M_2^*-\big(\overline M_1^*\big)^2\Big)h^2$$

$$\displaystyle \overline m_3 = \Big(\overline M_3^* - 3\cdot \overline M_2^*M_1^* + 2\cdot \big(\overline M_1^*\big)^3\Big)h^3$$

$$\displaystyle \overline m_4 = \Big(\overline M_4^* - 4 \overline M_3^*M_1^* +6M_2^*\big(\overline M_1^*\big)^2-3\big(\overline M_1^*\big)^4\Big)h_4$$

$\bullet$ Мода интервального ряда вычисляется как
$$M_0 = x_0 + \dfrac{n_M - n_{M-1}}{(n_M - n_{M-1}) + (n_M - n_{M+1})} * h$$
где $x_0$ - нижнаяя граница модального интервала, $h$ - длина модального интервала, $n_M$ - частота модального интервала

$\bullet$ Медиана интервального ряда:
$$m_e = x_0 + \dfrac{0.5n - n_{m-1}^н}{n_m} * h$$
где $x_0$ - нижнаяя граница медианного интервала, $h$ - длина медианного интервала, $n_m$ - частота медианного интервала
