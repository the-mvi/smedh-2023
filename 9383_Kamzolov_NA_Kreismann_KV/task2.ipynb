## Основные теоретические требования

*Статистическое оценкой* $\Theta^*$ неизвестного параметра теоретического распределения $\Theta$ называется функция он наблюдаемых значений случайной величины:

$$\Theta^* = f(x_1,x_2,...,x_n)$$

Статистические оценки, определяемые одним числом, называются *точечными*.

Оценка называется *несмещенной*, если ее математическое ожидание равно оцениваемому параметру $\Theta$ при любом объеме выборке.
$$M(\Theta^*) = \Theta$$

Оценка $\Theta^*$ называется *эффективной*, если при заданном объеме выборки n она имеет наименьшую возможную дисперсию, т.е.:
$$D(\Theta^*)\rightarrow min$$

Оценка $\Theta^*$ называется *состоятельной* для параметра $\Theta$, если она сходится по вероятности к оцениваемому параметру при неограниченном возрастании объема выборке, т.е.:
$$\lim_{N\to \infty} P(|\Theta - \Theta^*|\leq\varepsilon) =1$$,
где $\varepsilon >0 $ - сколькоугодно малое положительное число.

Начальным эмпирическим моментом k-го порядка называется среднее значение k-х степенй элементов вариационного или интервального ряда.
$$\overline{M_k} = \frac{1}{N} \sum{n_j x_j^k}$$,
$n_j$ - абсолютная частота j-ой варианты, $x_j$ - j-ая варианта, N - объем выборки.

Выборочное среднее вычисляется по формуле:

$$\overline{x_в} = \overline{M_1} = \frac{1}{N} \sum n_j x_j$$

Центральным эмирическим моментом k-го порядка называется среднее значение k-х степеней разностей ($x_j - \overline{M_1}$) для вариационного или интервального ряда:
$$\overline{m_k} = \frac{1}{N} \sum{n_j(x_j - \overline{x_в})^k}$$

Выборочная дисперсия - центральный эмпирический момент второго порядка:
$$D_в = \overline{m_2} = \frac{1}{N} \sum{n_j (x_j - \overline{x_в})^2}$$

Исправленная оценка дисперсии:
$$s^2 = \frac{N}{N-1} D_в$$

Соответствующие среднеквадратические отклонения вычисляются как квадратный корень из соответствующих оценок дисперсии:
$$\sigma_в = \sqrt{D_в}; s = \sqrt{s^2}$$

Статистическая оценка асимметрии:
$$\overline{A_s} = \frac{\overline{m_3}}{s^3}$$

Статистическая оценка эксцесса:
$$\overline{E} = \frac{\overline{m_4}}{s^4} - 3$$

Для упрощения вычислений центральных эмпирических моментов вводятся условные эмпирические моменты и условные эмирические варианты.

Условными вариантами называются варианты, определенные равенством: 
$$u_j = \frac{x_j-C}{h}$$,
где C - условный ноль, значение которого выбирается равным значению варианты интервального ряда, являющейся средней или близкой к средней по значению в этом ряду ; h - длина интервала в интервальном ряду.

Условным эмпирическим моментом порядка k называется начальный момент порядка k, вычисленный для условных вариант:
$$\overline{M_k^*} = \frac{1}{N} \sum n_j (\frac{x_j-C}{h})^k = \frac{1}{N} \sum{n_j u_j^k}$$


В частности:

$$\overline{M_1^*} = \frac{1}{N} \sum n_j (\frac{x_j-C}{h}) = \frac{1}{N} \sum{n_j u_j}$$

Условные эмпирические моменты связаны с эмпирическими начальными и центральными моментами следующим образом:
$$\overline{x}_в = \overline{M}_1 = \overline{M}_1^* h + C$$
$$\overline{m}_2 = (\overline{M}_2^* - (\overline{M}_1^*)^2)h^2$$
$$\overline{m}_3 = (\overline{M}_3^* - 3\overline{M}_2^*\overline{M}_1^* + 2(\overline{M}_1^*)^3)h^3$$
$$\overline{m}_4 = (\overline{M}_4^* - 4\overline{M}_3^*\overline{M}_1^* + 6\overline{M}_2^*(\overline{M}_1^*)^2 - 3(\overline{M}_1^*)^4)h^4$$


Модой $M_o^*$ вариационного ряда называется варианта, которой соответствует наибольшая частота.

Для интервального ряда мода находится внутри интервала ($x_{i-1},x_i$), которому соответствует наибольшая частота и ее приближенное значение может быть вычислено по формуле:

$$M_o^* = x_{i-1}+h\frac{\tilde{m}_i - \tilde{m}_{i-1}}{(\tilde{m}_i - \tilde{m}_{i-1})+(\tilde{m}_i - \tilde{m}_{i+1})}$$
, где h - длина интервала; $\tilde{m}_i $ - относительная частота для i-го интервала.

Медианой $M_e^*$ называется варианта, соответствующая середине ранжированного ряда.

Если ранжированный ряд имеет нечетное (2m+1) количество элементов, то $M_e^* = x_{m+1}$.

Если ранжированный ряд имеет четное (2m) количество элементов, то $M_e^* = \frac{(x_m + x_{m+1})}{2}$

Для интервального ряда медиана вычисляется по формуле:
$$M_e^* = x_{i-1} + \frac{h}{\tilde{m}_i}(0.5 - \sum_{j=1}^{i-1}\frac{h}{\tilde{m}_j})$$


Выборочный коэффициент вариации $V^*$ характеризует разброс значений вариант относительно выборочного среднего:
$$V^* = \frac{\sigma_в}{\overline{x}_в}*100\%$$

#### Вычисление условных вариант для середин интервального ряда

```python
left_bounds = [344,379,414,449,484,519,554]#левые границы интервалов
middle_points = [361.5, 396.5, 431.5, 466.5, 501.5, 536.5, 569.5] #середина i-го частичного интервала
absolute_freq = [11,13,35,20,15,9,4]#абсолютная частота i-го частичного интервала
relative_freq = [0.102804,0.121495,0.327103,0.186916,0.140187,0.0841121,0.0373832]#относительная частота i-го интервала
accum_relative_freq = [0.102804,0.224299,0.551402,0.738318,0.878505,0.962617,1]#накопленная частота для i-го интервала
C = middle_points[3]#ложный ноль
```

```python
u_vars = [(i-C)/h for i in middle_points]#условные варианты
u_vars
```

```python
def add_sum(lst):#добавляет элемент в список и возвращает список
    res = lst.copy()
    res.append(sum(res))
    return res
```

#### Вычисление эмпирических моментов

```python
#запишем суммы в переменные
sum_n_u = sum(n_u)
sum_n_u2 = sum(n_u2)
sum_n_u3 = sum(n_u3)
sum_n_u4 = sum(n_u4)

#Посчитаем условные эмпирические моменты
nu1 = sum_n_u/N
nu2 = sum_n_u2/N
nu3 = sum_n_u3/N
nu4 = sum_n_u4/N

#Посчитаем центральные эмпирические моменты
mu1 = nu1*h+C
mu2 = (nu2 - nu1**2)*h**2
mu3 = (nu3 - 3*nu1*nu2 +2*nu1**3)*h**3
mu4 = (nu4 - 4*nu3*nu1 + 6*nu2*nu1**2 -3*nu1**4)*h**4

table2 = pd.DataFrame(
    {
        "$\\nu_i^*$":[nu1,nu2,nu3,nu4],
        "$\mu_i^*$":[mu1,mu2,mu3,mu4]
    }
)#сформированная таблица
table2.index = table2.index + 1
print(table2.to_markdown())
```

#### Вычисление выборочной средней и дисперсии

```python
#вычисление стандартным способом выборочного среднего
standart_mean = sum([absolute_freq[i]*middle_points[i] for i in range(k)])/N
#сравнение с вычисленным с помощью условных вариант
print(standart_mean, standart_mean == mu1)
```

```python
#вычисление стандартным способом выборочной дисперсии
var = sum([absolute_freq[i]*(middle_points[i]-standart_mean)**2 for i in range(k)])/N
#сравнение с вычисленным с помощью условных вариант
print(var, var == mu2)
```

#### Вычисление исправленной выборочной дисперсии и СКО.

```python
s2 = var*N/(N-1)#исправленная выборочная дисперсия
s2
```

```python
s  = math.sqrt(s2)#исправленное СКО
s
```

```python
sigma = math.sqrt(var)#смещенная СКО
sigma
```

#### Вычисление моды, медианы и коэффициента вариации


```python
index_mode = absolute_freq.index(max(absolute_freq))#индекс интервала с наибольшей частотой

mode = left_bounds[index_mode] + h*(relative_freq[index_mode] - relative_freq[index_mode-1])/\
    ((relative_freq[index_mode] - relative_freq[index_mode-1])+(relative_freq[index_mode]-relative_freq[index_mode+1]))
#приближенная мода, вычисленная для интервального ряда
mode
```

```python
index_median = -1 #индекс интервала i, для которого накопленная частота >=0.5, а для интервала i-1 меньше 0.5
for i in range(len(accum_relative_freq)):
    if accum_relative_freq[i]>=0.5:
        index_median = i
        break

median =  left_bounds[index_median] +\
    (h/relative_freq[index_median])*(0.5-accum_relative_freq[index_median-1])
median #приближенная медиана, вычисленная для интервального ряда
```
