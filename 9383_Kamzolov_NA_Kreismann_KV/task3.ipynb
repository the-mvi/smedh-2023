## Основные теоретические положения

**Интервальной** называется статистическая оценка, которая задается двумя числами - концами интервала (в котором может находиться значение параметра $\Theta$)

**Точностью $\Theta^*$** - статистической оценки параметра $\Theta$ называется значение $\delta$, удоавлетворяющее неравенству: 
<p style="font-size: 14px; text-align: center;"> $\delta > |\Theta^* - \Theta|$ </p>

**Надежностью (доверительной вероятностью)** оценки $\Theta^*$ называется вероятность $\gamma$, с которой выполняется неравенство:
<p style="font-size: 14px; text-align: center;"> $\gamma = P\{\delta > |\Theta^* - \Theta|\}$ </p>

Интервал $(\Theta^* - \delta; \Theta^* + \delta)$ называется **доверительным интервалом** и является интервальной оценкой параметра $\Theta$

Значение $q = 1 - \gamma$ называют уровнем значимости.

**Интервальной оценкой математического ожидания при известном СКВО** служит доверительный интервал:
<p style="font-size: 14px; text-align: center;"> $\overline{x}_s - t\sigma/\sqrt N < a < \overline{x}_s + t\sigma/\sqrt N$, где </p>
$\overline{x}_s$ - выборочное среднеее - статистическая оценка математического ожидания.

$\sigma$ - среднекватратичное отклонение.

$N$ - размер выборки

$t = \delta\sqrt(N)/\sigma$

Уровень значимости можно посчитать с помощью функции Лапласа $\Phi$ (имеются табличные значения):
<p style="font-size: 14px; text-align: center;"> $P\{\overline{x}_s - t\sigma/\sqrt N < a < \overline{x}_s + t\sigma/\sqrt N\} = 2\Phi(t) = \gamma$ </p>

**Интервальной оценкой математического ожидания при неизвестном СКВО** служит доверительный интервал:

<p style="font-size: 14px; text-align: center;"> $\overline{x}_s - t_{\gamma}S/\sqrt N < a < \overline{x}_s + t_{\gamma}S/\sqrt N$, где </p>

$\overline{x}_s$ - выборочное среднеее - статистическая оценка математического ожидания.

$S$ - исправленное выборочное СКВО.

$N$ - размер выборки

$t_\gamma$ - может быть определено по заданным значениям $\gamma$ и $N$ по таблице.


**Интервальной оценкой среднего квадратического отклонения** служит неравенство:
<p style="font-size: 14px; text-align: center;"> $S(1-q) < \sigma < S(1+q)$, где </p>
$S$ - исправленное выборочное СКВО

$q$ - может быть определено по заданным значениям $\gamma$ и $N$ по таблице.

**Статистической гипотезой** называют гипотезу о виде неизвестного распределения исследуемой случайной величины или о значении параметров известных распределений.

Выдвинутую гипотезу называют **нулевой или основной гипотезой и обозначают $H_0$**

**Конкурирующей или альтернативной** называют гипотезу $H_1$, которая противоречит основной гипотезе.

**Ошибка первого рода** состоит в том, что нулевая гипотеза отвергается, хотя она верна.

**Ошибка второго рода** состоит в том, что нулевая гипотеза принимается, хотя она неверна.

Вероятность совершения ошибки первого рода называют **уровнем значимости** и обозначают через $\alpha$

Величину $1-\beta$ - вероятность того, что гипотеза отвергается, когда она неверна называется **мощностью критерия**

В качестве провекри гипотезы можно использовать критерий **$\chi^2$(Хи-квадрат) или иначе - критерий Пирсона.**

Согласно этому критерию на основании выборочных данных вычисляется "наблюденное" значение случайной величины $\chi^2$:

<p style="font-size: 14px; text-align: center;"> $\chi^2_{obs} = \sum_{i=1}^K \frac{(n_i - n_i^{'})^2}{n_i^{'}}$</p>

Здесь $n_i^{'}$ - теоретические частоты - частоты, с которыми нормально распределенная величина с параметрами $\overline{x}_s$ и $S$ попадала бы в $i$-ый интервал интервального ряда при проведении $N$ испытаний.

Распределение хи-квадрат зависит от числа **степеней свободы $k$**, которая в данном случае вычисляется по формуле $k = K - 3$, K - количество интервалов в интервальном ряде.

По числу степеней свободы и уровню значимости **вычисляется критическое значение хи-квадрат** $\chi^2_{crit}=\chi^2(\alpha, k)$. Значение можно определить из таблиц. Область принятия гипотезы $H_0$ определяется условием:
<p style="font-size: 14px; text-align: center;"> $\chi^2_{obs} < \chi^2_{crit}$ </p>

Для **вычисления "теоретических" частот** можно воспользоваться следующей последовательностью действий:
1. Пересчитываем значения границ интервалов интервального ряда. При этом левая граница первого интервала назначается равной $-\inf$, а правую границу последнего интервала назначаем равной $+\inf$. Остальные значения вычисляем по формуле:
<p style="font-size: 14px; text-align: center;"> $z_i = \frac{x_i - \overline{x}_s}{S}, i = 2, 3, ..., K$, где </p>
$x_i$ - граница интервала
2. Вычисляем вероятности попадания случайной величины в каждый $i$-ый инетрвал интервального ряда:
<p style="font-size: 14px; text-align: center;"> $p_i = \Phi(z_{i+1}) - \Phi(z_{i}), i = 1, 2, 3, ..., K$ </p>
3. Вычислить значения "теоретических" частот: 
<p style="font-size: 14px; text-align: center;"> $n_i^{'}=N*p_i, i = 1, 2, 3, ..., K$ </p>
