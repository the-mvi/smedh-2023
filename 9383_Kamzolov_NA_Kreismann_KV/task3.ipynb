## –û—Å–Ω–æ–≤–Ω—ã–µ —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏–µ –ø–æ–ª–æ–∂–µ–Ω–∏—è

**–ò–Ω—Ç–µ—Ä–≤–∞–ª—å–Ω–æ–π** –Ω–∞–∑—ã–≤–∞–µ—Ç—Å—è —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∞—è –æ—Ü–µ–Ω–∫–∞, –∫–æ—Ç–æ—Ä–∞—è –∑–∞–¥–∞–µ—Ç—Å—è –¥–≤—É–º—è —á–∏—Å–ª–∞–º–∏ - –∫–æ–Ω—Ü–∞–º–∏ –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞ (–≤ –∫–æ—Ç–æ—Ä–æ–º –º–æ–∂–µ—Ç –Ω–∞—Ö–æ–¥–∏—Ç—å—Å—è –∑–Ω–∞—á–µ–Ω–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ $\Theta$)

**–¢–æ—á–Ω–æ—Å—Ç—å—é $\Theta^*$** - —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–π –æ—Ü–µ–Ω–∫–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ $\Theta$ –Ω–∞–∑—ã–≤–∞–µ—Ç—Å—è –∑–Ω–∞—á–µ–Ω–∏–µ $\delta$, —É–¥–æ–∞–≤–ª–µ—Ç–≤–æ—Ä—è—é—â–µ–µ –Ω–µ—Ä–∞–≤–µ–Ω—Å—Ç–≤—É: 
<p style="font-size: 14px; text-align: center;"> $\delta > |\Theta^* - \Theta|$ </p>

**–ù–∞–¥–µ–∂–Ω–æ—Å—Ç—å—é (–¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω–æ–π –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å—é)** –æ—Ü–µ–Ω–∫–∏ $\Theta^*$ –Ω–∞–∑—ã–≤–∞–µ—Ç—Å—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å $\gamma$, —Å –∫–æ—Ç–æ—Ä–æ–π –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è –Ω–µ—Ä–∞–≤–µ–Ω—Å—Ç–≤–æ:
<p style="font-size: 14px; text-align: center;"> $\gamma = P\{\delta > |\Theta^* - \Theta|\}$ </p>

–ò–Ω—Ç–µ—Ä–≤–∞–ª $(\Theta^* - \delta; \Theta^* + \delta)$ –Ω–∞–∑—ã–≤–∞–µ—Ç—Å—è **–¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–º –∏–Ω—Ç–µ—Ä–≤–∞–ª–æ–º** –∏ —è–≤–ª—è–µ—Ç—Å—è –∏–Ω—Ç–µ—Ä–≤–∞–ª—å–Ω–æ–π –æ—Ü–µ–Ω–∫–æ–π –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ $\Theta$

–ó–Ω–∞—á–µ–Ω–∏–µ $q = 1 - \gamma$ –Ω–∞–∑—ã–≤–∞—é—Ç —É—Ä–æ–≤–Ω–µ–º –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏.

**–ò–Ω—Ç–µ—Ä–≤–∞–ª—å–Ω–æ–π –æ—Ü–µ–Ω–∫–æ–π –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–≥–æ –æ–∂–∏–¥–∞–Ω–∏—è –ø—Ä–∏ –∏–∑–≤–µ—Å—Ç–Ω–æ–º –°–ö–í–û** —Å–ª—É–∂–∏—Ç –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª:
<p style="font-size: 14px; text-align: center;"> $\overline{x}_s - t\sigma/\sqrt N < a < \overline{x}_s + t\sigma/\sqrt N$, –≥–¥–µ </p>
$\overline{x}_s$ - –≤—ã–±–æ—Ä–æ—á–Ω–æ–µ —Å—Ä–µ–¥–Ω–µ–µ–µ - —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∞—è –æ—Ü–µ–Ω–∫–∞ –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–≥–æ –æ–∂–∏–¥–∞–Ω–∏—è.

$\sigma$ - —Å—Ä–µ–¥–Ω–µ–∫–≤–∞—Ç—Ä–∞—Ç–∏—á–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ.

$N$ - —Ä–∞–∑–º–µ—Ä –≤—ã–±–æ—Ä–∫–∏

$t = \delta\sqrt(N)/\sigma$

–£—Ä–æ–≤–µ–Ω—å –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ –º–æ–∂–Ω–æ –ø–æ—Å—á–∏—Ç–∞—Ç—å —Å –ø–æ–º–æ—â—å—é —Ñ—É–Ω–∫—Ü–∏–∏ –õ–∞–ø–ª–∞—Å–∞ $\Phi$ (–∏–º–µ—é—Ç—Å—è —Ç–∞–±–ª–∏—á–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è):
<p style="font-size: 14px; text-align: center;"> $P\{\overline{x}_s - t\sigma/\sqrt N < a < \overline{x}_s + t\sigma/\sqrt N\} = 2\Phi(t) = \gamma$ </p>

**–ò–Ω—Ç–µ—Ä–≤–∞–ª—å–Ω–æ–π –æ—Ü–µ–Ω–∫–æ–π –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–≥–æ –æ–∂–∏–¥–∞–Ω–∏—è –ø—Ä–∏ –Ω–µ–∏–∑–≤–µ—Å—Ç–Ω–æ–º –°–ö–í–û** —Å–ª—É–∂–∏—Ç –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª:

<p style="font-size: 14px; text-align: center;"> $\overline{x}_s - t_{\gamma}S/\sqrt N < a < \overline{x}_s + t_{\gamma}S/\sqrt N$, –≥–¥–µ </p>

$\overline{x}_s$ - –≤—ã–±–æ—Ä–æ—á–Ω–æ–µ —Å—Ä–µ–¥–Ω–µ–µ–µ - —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∞—è –æ—Ü–µ–Ω–∫–∞ –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–≥–æ –æ–∂–∏–¥–∞–Ω–∏—è.

$S$ - –∏—Å–ø—Ä–∞–≤–ª–µ–Ω–Ω–æ–µ –≤—ã–±–æ—Ä–æ—á–Ω–æ–µ –°–ö–í–û.

$N$ - —Ä–∞–∑–º–µ—Ä –≤—ã–±–æ—Ä–∫–∏

$t_\gamma$ - –º–æ–∂–µ—Ç –±—ã—Ç—å –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–æ –ø–æ –∑–∞–¥–∞–Ω–Ω—ã–º –∑–Ω–∞—á–µ–Ω–∏—è–º $\gamma$ –∏ $N$ –ø–æ —Ç–∞–±–ª–∏—Ü–µ.


**–ò–Ω—Ç–µ—Ä–≤–∞–ª—å–Ω–æ–π –æ—Ü–µ–Ω–∫–æ–π —Å—Ä–µ–¥–Ω–µ–≥–æ –∫–≤–∞–¥—Ä–∞—Ç–∏—á–µ—Å–∫–æ–≥–æ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è** —Å–ª—É–∂–∏—Ç –Ω–µ—Ä–∞–≤–µ–Ω—Å—Ç–≤–æ:
<p style="font-size: 14px; text-align: center;"> $S(1-q) < \sigma < S(1+q)$, –≥–¥–µ </p>
$S$ - –∏—Å–ø—Ä–∞–≤–ª–µ–Ω–Ω–æ–µ –≤—ã–±–æ—Ä–æ—á–Ω–æ–µ –°–ö–í–û

$q$ - –º–æ–∂–µ—Ç –±—ã—Ç—å –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–æ –ø–æ –∑–∞–¥–∞–Ω–Ω—ã–º –∑–Ω–∞—á–µ–Ω–∏—è–º $\gamma$ –∏ $N$ –ø–æ —Ç–∞–±–ª–∏—Ü–µ.

**–°—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–π –≥–∏–ø–æ—Ç–µ–∑–æ–π** –Ω–∞–∑—ã–≤–∞—é—Ç –≥–∏–ø–æ—Ç–µ–∑—É –æ –≤–∏–¥–µ –Ω–µ–∏–∑–≤–µ—Å—Ç–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –∏—Å—Å–ª–µ–¥—É–µ–º–æ–π —Å–ª—É—á–∞–π–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω—ã –∏–ª–∏ –æ –∑–Ω–∞—á–µ–Ω–∏–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –∏–∑–≤–µ—Å—Ç–Ω—ã—Ö —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–π.

**–ö–æ–Ω–∫—É—Ä–∏—Ä—É—é—â–µ–π –∏–ª–∏ –∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–æ–π** –Ω–∞–∑—ã–≤–∞—é—Ç –≥–∏–ø–æ—Ç–µ–∑—É $H_1$, –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–æ—Ç–∏–≤–æ—Ä–µ—á–∏—Ç –æ—Å–Ω–æ–≤–Ω–æ–π –≥–∏–ø–æ—Ç–µ–∑–µ.

**–û—à–∏–±–∫–∞ –ø–µ—Ä–≤–æ–≥–æ —Ä–æ–¥–∞** —Å–æ—Å—Ç–æ–∏—Ç –≤ —Ç–æ–º, —á—Ç–æ –Ω—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ –æ—Ç–≤–µ—Ä–≥–∞–µ—Ç—Å—è, —Ö–æ—Ç—è –æ–Ω–∞ –≤–µ—Ä–Ω–∞.

**–û—à–∏–±–∫–∞ –≤—Ç–æ—Ä–æ–≥–æ —Ä–æ–¥–∞** —Å–æ—Å—Ç–æ–∏—Ç –≤ —Ç–æ–º, —á—Ç–æ –Ω—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ –ø—Ä–∏–Ω–∏–º–∞–µ—Ç—Å—è, —Ö–æ—Ç—è –æ–Ω–∞ –Ω–µ–≤–µ—Ä–Ω–∞.
